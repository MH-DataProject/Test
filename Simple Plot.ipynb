# Simple Example Of Lineare Regression Plot

## Creating Data Vectors

```r
# number of vector elements
n <- 40

# create a vector of integers
x <- seq(n)

set.seed(9485)
         
# create a vector of normal distributes values
rnorm_vec <- rnorm(n, mean=0, sd=5)

# create y
y <- x- rnorm_vec
```

## Creating Plots

```r
# histogram of y
hist(y)
```

```r
# x vs. y
plot(x, y, pch = 16, cex = 1.0, col = "blue", main = "Linear Regression", xlab = "x", ylab = "y")
# linear regression - trend line
abline(lm(y ~ x))
```
